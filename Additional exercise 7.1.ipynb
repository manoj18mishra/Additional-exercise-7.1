```python
import pandas as pd
from sklearn.datasets import load_iris 
iris = load_iris() 
data = pd. DataFrame(iris.data, columns=iris.feature_names) 
label = pd.DataFrame(list(map(lambda x : iris.target_names[x], iris.target)), columns=['Species'])
iris = pd.concat([data, label], axis=1) 
print(iris.head())
```

### 1) Use the distplot() to see the distribution of the SepalLengthCm, SepalWidthcm, P etalLengthCm, PetalWidthCm features. Plot them as subplots in a single image.

```python
import seaborn as sb
import matplotlib.pyplot as plt
sb.set(style="white", palette="muted", color_codes=True)
f, axes = plt.subplots(2, 2)
sb.distplot(iris["sepal length (cm)"],color='b',ax=axes[0,0])
sb.distplot(iris["sepal width (cm)"],color='r',ax=axes[0,1])
sb.distplot(iris["petal length (cm)"],color='g',ax=axes[1,0])
sb.distplot(iris["petal width (cm)"],color='m',ax=axes[1,1])
```

### 2) Do a boxplot of all features except 'Species'.

```python
sb.boxplot(data=iris.loc[:,iris.columns!='Species'])
```

### 3) Do a countplot for the feature 'Species'

```python
sb.countplot(iris['Species'])
```

### 4) Do a pairplot on the features 'SepalLengthcm', 'SepalWidthcm', 'PetalLengthCm','PetalWidthCm', 'Species'.


```python
sb.pairplot(iris)
```

### 5) Do an Implot on the following SepalLengthcm, PetalLengthcm. Using hue, display the different species in different colors.

```python
sb.lmplot("sepal length (cm)","petal length (cm)",data=iris,hue='Species')
```

### 6) Do a barplot of 'Species' vs 'SepalLengthCm'.

```python
sb.barplot('Species','sepal length (cm)',data=iris)
```

### 7) Using heatmap, plot the correlation matrix.

```python
sb.heatmap(data=iris.corr())
```
